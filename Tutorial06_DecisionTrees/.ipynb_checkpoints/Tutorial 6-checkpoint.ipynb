## Decision Trees

This week, we're going to learn about **Decision Trees** an algorithm which classifys a data point based on that point's features. The decision tree works by answering a sereis of yes / no questions about the data point's features. The result of these questions is a classificaiton. 

In the below example, we want to decide whether or not the day is a good day for baseball ("Siri, is it a good day to play baseball?"). We have acess to a few bits of information.

### Building the Tree

The first step in building a tree is to select a feature to split on. How do we do this? It turns out there are a different ways which give slightly different results. Before we look at thoes, let's think about what would make a good split...

<img src="split.png">

The most common is splitting based on **Information Gain**. Which makes use of a metric called **Shannon Entropy**

$$
Entropy = H(y) = -\sum_{y \in Y} P(y)*log_2(P(y))
$$

$$
Conditional Entropy = H(Y|X) = \sum_{x \in X} P(X=x)*H(Y|X=x)
$$

$$
Information Gain = IG(x,y) = H(Y) - H(Y|X)
$$

Intuitively, **entropy** can be thought of as the *peakiness* of the histogram. If the histogram is very flat, than the variable has large entropy, if the variable is very skewed, than there is less entropy. **Conditional Entropy** is the entropy in the presence of some new information, in this case, the value of $X$ on which we might want to split. Therefore, **Information Gain** is the decrease in entropy associated with a given feature. 

So what does this have to do with building a tree? Well, we take the fature with the greatest informaiton gain, this let's us build an optimal tree with the fewest number of branches. 

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from scipy import stats
import pylab as pl

import fig_code
```

## Random Forrests

Random forrests works by introducing a few random elements to collectionfo of decision trees. The algorithm works as follows...

We first build $k$ trees, and train them on a bootstrapped samples of data. Instead of splitting on all the features, we select a random subset of features to split on. We then calculate information gain as before. And build all the trees (a forest).

Once the trees are build, data is classified by all the trees in the forrest, and then the final classificaiton is decided by majority vote. 

```python
def fit_randomized_tree(random_state=0):
    X, y = make_blobs(n_samples=300, centers=4,
                      random_state=0, cluster_std=2.0)
    clf = DecisionTreeClassifier(max_depth=15)
    
    rng = np.random.RandomState(random_state)
    i = np.arange(len(y))
    rng.shuffle(i)
    visualize_tree(clf, X[i[:250]], y[i[:250]], boundaries=False,
                   xlim=(X[:, 0].min(), X[:, 0].max()),
                   ylim=(X[:, 1].min(), X[:, 1].max()))
    
from IPython.html.widgets import interact
interact(fit_randomized_tree, random_state=[0, 100]);
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
visualize_tree(clf, X, y, boundaries=False);
```

Comparing the images, we can see that the second one is much cleaner. Suggesting that the classifier is more generalizable. 

## Random Forest Limitations

The following data scenarios are not well suited for random forests:
* y: lots of 0, few 1 (class imbalances)
* Structured data like images where a neural network might be better
* Small data size which might lead to overfitting
* High dimensional data where a linear model might work better
